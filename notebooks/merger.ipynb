```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars1 = pd.read_csv("../data/raw/cars1.csv")
```

```python
cars2 = pd.read_csv("../data/raw/cars2.csv")
```

```python
scraped = pd.read_csv("../data/raw/cars_scraped.csv")
scraped.shape
```

```python
scraped.head()
```

```python
scraped.drop(columns=['price_text', 'search_city', 'listing_url', 'scraped_at', 'owner_slug', 'seller_type'], inplace=True)
scraped.head()
```

```python
scraped.shape
```

```python
cars1.head()
```

```python
cars1.drop(columns=['Color', 'Max Power', 'Max Torque', 'Drivetrain', 'Length', 'Width', 'Height', 'Fuel Tank Capacity', 'Seller Type'], inplace=True)
cars1.head()
```

```python
cars1.shape
```

```python
cars2.head()
```

```python
cars2.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'New_Price', 'Power', 'Mileage'], inplace=True)
cars2.head()
```

```python
cars2.shape
```

```python
cars1.info()
```

```python
import pandas as pd
import numpy as np

# Align columns
scraped.rename(columns={'actual_city': 'location'}, inplace=True)

# Cars2 price is in Lakhs → convert
cars2['Price'] = cars2['Price'] * 100_000

# Parse brand + model from Cars2 Name column
cars2['brand'] = cars2['Name'].str.split().str[0]
cars2['model'] = cars2['Name'].str.split(n=1).str[1]
cars2.drop(columns=['Name'], inplace=True)

# Normalize column names
cars1.rename(columns={
    'Make': 'brand', 'Model': 'model', 'Price': 'price',
    'Year': 'year', 'Kilometer': 'km', 'Fuel Type': 'fuel',
    'Transmission': 'transmission', 'Location': 'location',
    'Owner': 'owner', 'Seller Type': 'seller_type',
    'Seating Capacity': 'seats'
}, inplace=True)

cars2.rename(columns={
    'Price': 'price', 'Year': 'year', 'Kilometers_Driven': 'km',
    'Fuel_Type': 'fuel', 'Transmission': 'transmission',
    'Location': 'location', 'Owner_Type': 'owner', 'Seats': 'seats'
}, inplace=True)

df_merged = pd.concat([scraped, cars1, cars2], ignore_index=True)

df_merged.head()
```

```python
df_merged.info()
```

```python
# Check where price nulls come from
print(df_merged[df_merged['price'].isnull()]['brand'].value_counts().head(10))
```

```python
df_merged.isnull().sum()
```

```python
df_merged.drop(columns=['variant', 'body_type', 'locality'], inplace=True)
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged.dropna(inplace=True)
```

```python
df_merged.shape
```

```python
df_merged.rename(columns={
    'brand': 'Brand', 'model': 'Model', 'price': 'Price',
    'year': 'Year', 'km': 'Km', 'fuel': 'Fuel',
    'transmission': 'Transmission', 'location': 'Location',
    'owner': 'Owner','seats': 'Seats',
}, inplace=True)
```

```python
df_merged['Engine'] = df_merged['Engine'].str.extract('(\d+)').astype(int)
```

```python
df_merged.to_csv("../data/processed/cars_processed.csv", index=False)
```
